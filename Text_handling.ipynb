```python
import string
```

```python
with open('conv.txt') as f:
    lines = f.readlines()
    new_lines = []
    for i in range(len(lines)):
        if i % 2 == 0:
            new_lines.append(lines[i].strip())
new_lines
```

```python
actor_words = {}
```

```python
for line in new_lines:
    name = line.split(':')[0] # Actor Name
    try:
        actor_words[name] = {}
    except:
        pass
actor_words
```

```python
actor_words.keys()
```

```python
print("Number of unique dialogue speakers are : ",len(actor_words.keys()))  #Task-1 completed
```

<h4>Task-1 Completed<h4>

```python
for line in new_lines:
    split = line.split(':')
    actor = split[0]
    temp_words = split[1].split(' ')[1:]
    
    for i in range(len(temp_words)):
        word = temp_words[i]
        word = word.lower()
        word = word.replace('â€™', '\'')
        temp_words[i] = word
        if word != 'â€¦':
            word = word.replace(',', '')
            word = word.replace('.', '')
            word = word.replace('?', '')
            word = word.replace('!', '')
            for a in actor_words.keys():
                if word not in actor_words[a].keys():
                    actor_words[a][word] = 0
        
            actor_words[actor][word] += 1
actor_words
```

```python
for actor in actor_words.keys():
    print(actor, '\n')
    for word in actor_words[actor].keys():
        print(f'{word}: {actor_words[actor][word]}')
    print()
```

```python
for actor in actor_words.keys():
    unique_words = set()
    for word in actor_words[actor].keys():
        word = word.lower()
        if actor_words[actor][word] > 0:
            for temp_actor in actor_words.keys():
                if temp_actor == actor:
                    continue
                if actor_words[temp_actor][word] == 0:
                    unique_words.add(word)
    with open(f'{actor}.txt', 'w') as f:             #Generating text files
        for word in unique_words:                     
            f.write(f'{word}\n')
        print(f.name + " file is generated and words stored successfully.")
```
